Construa um gráfico da Váriavel Temperatura Média

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_table('../Dados/Consumo_cerveja.csv', sep=';')
df.head()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(16,8)
ax.set_title('Temperatura Média', fontsize=16)
ax.set_ylabel('°C', fontsize=16)
ax = df['temp_media'].plot()
```

Construa o boxplot do consumo por final de semana ou não, com consumo na vertical

```python
import seaborn as sns
```

```python
sns.set_palette('icefire')
sns.set_style('darkgrid')
ax = sns.boxplot(x='fds', y='consumo', data=df, orient='v')
ax.figure.set_size_inches(12,6)
ax.set_xlabel('Final de Semana')
ax.set_ylabel('Litros')
ax.set_title('Consumo por Final de Semana')
```
